```python
# ============================================================
# HYBRID FRAMEWORK EVALUATION
# ============================================================

import pandas as pd

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)

from injex.predictor import get_score
from injex.decision_engine import decide

# ============================================================
# LOAD EXTERNAL DATASET
# ============================================================

df = pd.read_csv(
    r"D:\Prompt_Injection\dataset\test_updated.csv"
)

# ============================================================
# CHECK COLUMNS
# ============================================================

print("\nDATASET COLUMNS:\n")

print(df.columns)

# ============================================================
# STORE RESULTS
# ============================================================

y_true = []

y_pred = []

# ============================================================
# RUN HYBRID FRAMEWORK
# ============================================================

for _, row in df.iterrows():

    # input text
    text = row["text"]

    # actual label
    actual_label = row["label"]

    # ========================================================
    # GET HYBRID SCORES
    # ========================================================

    scores = get_score(text)

    # final hybrid score
    final_score = scores["final_score"]

    # final decision
    decision = decide(final_score)

    # ========================================================
    # CONVERT DECISION TO LABEL
    # ========================================================

    # BLOCK / REVIEW => malicious
    if decision in ["BLOCK", "REVIEW"]:

        predicted_label = 1

    # ALLOW => safe
    else:

        predicted_label = 0

    # store labels
    y_true.append(actual_label)

    y_pred.append(predicted_label)

# ============================================================
# CALCULATE METRICS
# ============================================================

accuracy = accuracy_score(
    y_true,
    y_pred
)

precision = precision_score(
    y_true,
    y_pred
)

recall = recall_score(
    y_true,
    y_pred
)

f1 = f1_score(
    y_true,
    y_pred
)

# ============================================================
# PRINT RESULTS
# ============================================================

print("\n" + "=" * 60)

print("HYBRID FRAMEWORK RESULTS")

print("=" * 60)

print(f"\nAccuracy : {accuracy:.4f}")

print(f"Precision: {precision:.4f}")

print(f"Recall   : {recall:.4f}")

print(f"F1 Score : {f1:.4f}")

# ============================================================
# CONFUSION MATRIX
# ============================================================

cm = confusion_matrix(
    y_true,
    y_pred
)

print("\n" + "=" * 60)

print("CONFUSION MATRIX")

print("=" * 60)

print("\n")

print(cm)

# ============================================================
# CLASSIFICATION REPORT
# ============================================================

print("\n" + "=" * 60)

print("CLASSIFICATION REPORT")

print("=" * 60)

print("\n")

print(
    classification_report(
        y_true,
        y_pred
    )
)
```
