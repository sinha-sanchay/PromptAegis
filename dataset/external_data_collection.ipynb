Downloading dataset to local space

```python
from datasets import load_dataset

# Load dataset from Hugging Face
ds = load_dataset("jayavibhav/prompt-injection-safety")

# Convert train split to pandas
train_df = ds["train"].to_pandas()

# Save to CSV
train_df.to_csv("train.csv", index=False)

# Convert test split to pandas
test_df = ds["test"].to_pandas()

# Save to CSV
test_df.to_csv("test.csv", index=False)

print("Dataset saved successfully.")
```

```python
import pandas as pd

train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")

print("Train rows:", len(train))
print(train["label"].value_counts())

print("\nTest rows:", len(test))
print(test["label"].value_counts())
```

```python
# convert label 2 -> label 1
train["label"] = train["label"].replace(2, 1)
test["label"] = test["label"].replace(2, 1)

# save updated files
train.to_csv("train_updated.csv", index=False)
test.to_csv("test_updated.csv", index=False)

print("Files saved as train_updated.csv and test_updated.csv")
```

```python

train_updated = pd.read_csv("train_updated.csv")
test_updated = pd.read_csv("test_updated.csv")

print("Train rows:", len(train_updated))
print(train_updated["label"].value_counts())

print("\nTest rows:", len(test_updated))
print(test_updated["label"].value_counts())
```
